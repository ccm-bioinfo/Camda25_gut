```python
import pandas as pd
```

```python
def paired_to_unpaired(index):
    one_sample_index=set()
    for sample in index:
        one_sample_index.add(sample[:-1])
    return list(one_sample_index)
```

```python
def unpaired_to_paired(index):
    split_sample=list()
    for sample in index:
        split_sample.append(sample+'a')
        split_sample.append(sample+'b')
    return split_sample
```

<h3> Bajo el supuesto de que un mismo diagnosis implica una menor distancia en la beta diversidad y diferente diagnosis implica mayor distancia, se construye el siguiente código. Se asigna para "cluster2" 1 si tiene poca distancia, y 2 si tiene gran distancia</h3> 

```python
hipca_predicted_cluster=pd.Series(index=beta_div.index,dtype='Int64')
for sample in beta_div.index:
    prediction_a_hipca=hipca.loc[sample+'a','Prediction']
    prediction_b_hipca=hipca.loc[sample+'b','Prediction']
    if prediction_a_hipca==prediction_b_hipca:
        hipca_predicted_cluster[sample]=1
    else:
        hipca_predicted_cluster[sample]=2
beta_div['hipca_cluster']=hipca_predicted_cluster
```

```python
RFT_predicted_cluster=pd.Series(index=beta_div.index,dtype='Int64')
for sample in beta_div.index:
    prediction_a_RFT=RFT.loc[sample+'a','RFT']
    prediction_b_RFT=RFT.loc[sample+'b','RFT']
    if prediction_a_RFT==prediction_b_RFT:
        RFT_predicted_cluster[sample]=1
    else:
        RFT_predicted_cluster[sample]=2
beta_div['RFT_cluster']=RFT_predicted_cluster
```
