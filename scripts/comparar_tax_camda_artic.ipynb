<h3>El conjunto diff_article_dict son los nombres de las muestras del articulo menos los del diccionario</h3>

```python
diff_article_dict=set(tax_article.columns).difference(train_to_sample.index)
```

```python
changed_article_sample_names={}
set_article_sample_names=set(tax_article.columns)
set_dict_sample_names=set(train_to_sample.index)
for sample in set_article_sample_names:
    if sample in set_dict_sample_names:
        changed_article_sample_names[sample]=sample
    elif sample.replace('-','.') in set_dict_sample_names:
        changed_article_sample_names[sample]=sample.replace('-','.')
```

```python
len(changed_article_sample_names.keys())
```

```python
for i in changed_article_sample_names.keys():
    if i!=changed_article_sample_names[i]:
        print(i,changed_article_sample_names[i])
```

```python
train_to_sample.loc[list(changed_article_sample_names.values()),]
```

```python
tax_article.rename(columns=changed_article_sample_names,inplace=True)
```

```python
tax_article_dep=tax_article.loc[list(set(tax_article.index).intersection(set(tax_camda.index))),train_to_sample.index]
tax_article_dep
```

```python
intersec_article_camda_index=list(set(tax_article_dep.index).intersection(set(tax_camda.index)))
```

```python
len(intersec_article_camda_index)
```

```python
tax_article_rest=tax_article_dep.loc[intersec_article_camda_index,]
tax_article_rest.sort_index(inplace=True)
tax_article_rest
```

```python
tax_camda_res=tax_camda.loc[intersec_article_camda_index,train_to_sample['Unnamed: 0']]
tax_camda_res.sort_index(inplace=True)
tax_camda_res
```

<h3>Ahora se revisa si son iguales los valores para cada muestra, es decir ver si coinciden las muestras propuestas del diccionario con las train de taxonomia</h3>

```python
for j in tax_article_rest.columns:
    trad_j=train_to_sample.loc[j]['Unnamed: 0']
    for i in tax_article_rest.index:
        if tax_article_rest.loc[i,j]!=tax_camda_res.loc[i,trad_j]:
            print(i)
```
